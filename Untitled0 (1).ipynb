```python
#@title ## **Sentences cleaning**
stop_words = stopwords.words('english')

# Converts the unicode file to ascii
def unicode_to_ascii(s):
    return ''.join(c for c in unicodedata.normalize('NFD', s)
        if unicodedata.category(c) != 'Mn')

def preprocess_sentence(w):
    w = unicode_to_ascii(w.lower().strip())
    # creating a space between a word and the punctuation following it
    # eg: "he is a boy." => "he is a boy ."
    w = re.sub(r"([?.!,¿])", r" \1 ", w)
    w = re.sub(r'[" "]+', " ", w)
    # replacing everything with space except (a-z, A-Z, ".", "?", "!", ",")
    w = re.sub(r"[^a-zA-Z?.!]+", " ", w)
    w = re.sub(r'\b\w{0,2}\b', '', w)

    #remove vocalsound
    w = w.replace("vocalsound", '')
    w = w.replace("um", '')
    w = w.replace("uh", '')
    w = w.replace("disfmarker", '')

    # remove stopword
    mots = word_tokenize(w.strip())
    # mots = [mot for mot in mots if mot not in stop_words]
    return ' '.join(mots).strip()
    return mots

def process_data(X):
  return [ preprocess_sentence(x) for x in X_training_original]
```

```python
#@title ## Training and Testing Files
def flatten(list_of_list):
    return [item for sublist in list_of_list for item in sublist]
#####
# training and test sets of transcription ids
#####
training_set = ['ES2002', 'ES2005', 'ES2006', 'ES2007', 'ES2008', 'ES2009', 'ES2010', 'ES2012', 'ES2013', 'ES2015', 'ES2016', 'IS1000', 'IS1001', 'IS1002', 'IS1003', 'IS1004', 'IS1005', 'IS1006', 'IS1007', 'TS3005', 'TS3008', 'TS3009', 'TS3010', 'TS3011', 'TS3012']
training_set = flatten([[m_id+s_id for s_id in 'abcd'] for m_id in training_set])
training_set.remove('IS1002a')
training_set.remove('IS1005d')
training_set.remove('TS3012c')

test_set = ['ES2003', 'ES2004', 'ES2011', 'ES2014', 'IS1008', 'IS1009', 'TS3003', 'TS3004', 'TS3006', 'TS3007']
test_set = flatten([[m_id+s_id for s_id in 'abcd'] for m_id in test_set])


#####
# text_baseline: utterances are embedded with SentenceTransformer, then train a classifier.
#####

y_training = []
graph = []
last_graph_size = 0
with open(root + "training_labels.json", "r") as file:
    training_labels = json.load(file)

X_training_original = []
for transcription_id in tqdm(training_set):
    with open(path_to_training / f"{transcription_id}.json", "r") as file:
        transcription = json.load(file)

    with open(path_to_training / f"{transcription_id}.txt", 'r') as file:
        relations = file.readlines()

    graph.extend([(lambda x: (int(x[0]) + last_graph_size, int(x[2]) + last_graph_size, x[1])) (rel.split(' ')) for rel in relations])

    last_graph_size += len(transcription)


    for utterance in transcription:
        X_training_original.append(utterance["speaker"] + ": " + utterance["text"])

    y_training += training_labels[transcription_id]
```

```python
class AlgorithmicModel:
  def __init__(self):
    pass
  def fit(self, X, y):
    X = np.array(X)
    y = np.array(y)
    important_sentences = X[y==1]
    random_sentences = X[y==0]
    iwords = []
    rwords = []
    for x in important_sentences:
      iwords.extend(x[4:].split(' '))
    for w in random_sentences:
      rwords.extend(w[4:].split(' '))
    idic = Counter(iwords)
    rdic = Counter(rwords)
    icount = np.array(sorted([(k, v) for k, v in idic.items()], key=lambda x: x[1], reverse=True))
    rcount = np.array(sorted([(k, v) for k, v in rdic.items()], key=lambda x: x[1], reverse=True))
    common = set(iwords) & set(rwords); len(common)/len(iwords)
    l = []
    for c in common:
      if idic[c] > rdic[c]:
        l.append((c, idic[c] / rdic[c]))
    l.sort(key=lambda x: x[1], reverse=True)
    l = np.array(l)

    self.vocab = l

  def predict(self, X):
    y = np.zeros(len(X))
    for i, x in enumerate(X):
      for l in self.vocab:
        # print(x); break
        if l[0] in x and float(l[1]) > 2.2:
          y[i] = 1
          break
      # if y[i] > 2:
      #   y[i] = 1
      # else:
      #   y[i] = 0
    return y

  def f1(self, X, y):
    return f1_score(y, self.predict(X))
```

```python
from tensorflow.keras import backend as K
def metric(y_true, y_pred):
    y_pred = K.round(y_pred)
    tp = K.sum(K.cast(y_true * y_pred, 'float'), axis=0)
    fp = K.sum(K.cast((1 - y_true) * y_pred, 'float'), axis=0)
    fn = K.sum(K.cast(y_true * (1 - y_pred), 'float'), axis=0)

    precision = tp / (tp + fp + K.epsilon())
    recall = tp / (tp + fn + K.epsilon())

    f1_score = 2 * precision * recall / (precision + recall + K.epsilon())

    return K.mean(f1_score)
```

```python
#@title ADABOOST
def train_and_predict_AdaBoost(B, X_train, y_train, X_test, y_test):
    """
    Inputs:
        B (int): number of boosting steps
        X_train (Nxp matrix): training data
        y_train (N vector): response training values
        X_test (N_test x p matrix): test data to be predicted
        y_test (N vector): response test values
    Outputs:
        acc_train (B vector): train accuracy per boosting step
        acc_test (B vector): test accuracy per boosting step
        w_all (NxB matrix): weights computed for all the samples per boosting step
    """
    # insert your code here
    N = X_train.shape[0]
    p = 0.01
    acc_train = np.zeros(B+1)
    acc_test = np.zeros(B+1)
    f1_scores = np.zeros(B+1)
    w_all = np.zeros((N, B))

    y_pred = np.zeros(N)
    y_pred_test = np.zeros(X_test.shape[0])

    sum_alpha = .0

    # 1. Initialize the weights
    w = np.ones(N) / N

    # 2. For b = 1, ..., B:
    for i in range(B):
        if i % 100 == 1:
            print(f"Step {i} : Train Accuracy : ", acc_train[i-1], "    |      Test Accuracy : ", acc_test[i-1])
        # 2.1 Fit a classifier with weights w
        Cb = DecisionTreeClassifier(max_depth=1)
        Cb.fit(X_train, y_train, sample_weight=w)

        pred = Cb.predict(X_train)

        # 2.2 Compute the weighted error eb
        I = pred != y_train
        eb = np.dot(w, I) / np.sum(w)

        # 2.3 Compute the coefficient alpha
        alpha_b = np.log((1 - eb) / eb)

        # 2.4 Update the weights
        w *= np.exp(alpha_b * I)

        largest = np.max(w)
        if w[w > largest / 10].size < N * p:
            w[w > largest / 10] = 2 * np.max(w[w < largest / 10])
        w_all[:, i] = w

        # 3. Predict the class of each sample in the training set and compute the train accuracy

        y_pred = ((y_pred * sum_alpha + pred * alpha_b) / (sum_alpha + alpha_b))
        acc_train[i+1] = np.mean(np.round(y_pred) == y_train)

        # 4. Predict the class of each sample in the test set and compute the test accuracy
        y_pred_test = ((y_pred_test * sum_alpha + Cb.predict(X_test) * alpha_b) / (sum_alpha + alpha_b))
        acc_test[i+1] = np.mean(np.round(y_pred_test) == y_test)

        f1_scores[i+1] = f1_score(y_test, np.round(y_pred_test))

        sum_alpha += alpha_b

    return acc_train[1:], acc_test[1:], w_all, f1_scores, y_pred_test

B = 500
acc_train, acc_test, w_all, f1_scores, y_pred = train_and_predict_AdaBoost(B, X_train, y_train, X_test, y_test)

print('The AdaBoost accuracy over the test sample is:', acc_test[B-1])

plt.figure(figsize=(15, 6))
plt.plot(acc_train, label="Training")
plt.plot(acc_test, label="Test")
plt.plot(f1_scores, label="F1-Score")
plt.xlabel('Boosting steps')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
def normalise_adjacency(A):
    #Please insert your code for Task 4 here

    # Calculate the degree matrix D
    D = np.diag(np.sum(A, axis=0).A1)

    # Add the identity matrix to A
    A_hat = A + sp.identity(A.shape[0])

    # Update D to reflect the added self-loops in A_hat
    D_hat = np.diag(np.sum(A_hat, axis=0).A1)

    # Calculate D_hat to the power of -0.5
    D_hat_inv_sqrt = sp.diags(np.power(D_hat.diagonal(), -0.5))

    # Return the normalised adjacency matrix
    return D_hat_inv_sqrt @ A_hat @ D_hat_inv_sqrt



def sparse_to_torch_sparse(M):
    """Converts a sparse SciPy matrix to a sparse PyTorch tensor"""
    M = M.tocoo().astype(np.float32)
    indices = torch.from_numpy(np.vstack((M.row, M.col)).astype(np.int64))
    values = torch.from_numpy(M.data)
    shape = torch.Size(M.shape)
    return torch.sparse.FloatTensor(indices, values, shape)
```

```python

class GCN(nn.Module):
    """Simple GCN model"""
    def __init__(self, n_feat, n_hidden_1, n_hidden_2, n_class, dropout):
        super(GCN, self).__init__()

        self.fc1 = nn.Linear(n_feat, n_hidden_1)
        self.fc2 = nn.Linear(n_hidden_1, n_hidden_2)
        self.fc3 = nn.Linear(n_hidden_2, n_class)
        self.dropout = nn.Dropout(dropout)
        self.relu = nn.ReLU()

    def forward(self, x_in, adj):

        #Please insert your code for Task 5 here
        # Layer 1
        z0 = self.fc1(x_in)
        z0 = torch.mm(adj, z0)  # Multiply with adjacency matrix
        z0 = self.relu(z0)
        z0 = self.dropout(z0)

        # Layer 2
        z1 = self.fc2(z0)
        z1 = torch.mm(adj, z1)  # Multiply with adjacency matrix
        z1 = self.relu(z1)

        # Output layer
        x = self.fc3(z1)
        return F.log_softmax(x, dim=1), z1
```

```python
# Hyperparameters
epochs = 200
n_hidden_1 = 512
n_hidden_2 = 128
learning_rate = 0.01
dropout_rate = 0.5

A = nx.adjacency_matrix(G)

# Creates the model and specifies the optimizer
model = GCN(features.shape[1], n_hidden_1, n_hidden_2, n_class, dropout_rate).to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

adj = normalise_adjacency(sp.csr_matrix(A))
adj = sparse_to_torch_sparse(sp.csr_matrix(A)).to(device)


def train(epoch):
    t = time.time()
    model.train()
    optimizer.zero_grad()
    output,_ = model(features, adj)
    loss_train = F.nll_loss(output[idx_train], class_labels_train)
    acc_train = accuracy_score(torch.argmax(output[idx_train], dim=1).detach().cpu().numpy(), class_labels_train.cpu().numpy())
    loss_train.backward()
    optimizer.step()


    model.eval()
    output,_ = model(features, adj)

    loss_val = F.nll_loss(output[idx_val], class_labels_validation)
    acc_val = f1_score(torch.argmax(output[idx_val], dim=1).detach().cpu().numpy(), class_labels_validation.cpu().numpy())
    if epoch%10 == 0:
        print('Epoch: {:03d}'.format(epoch+1),
              'loss_train: {:.4f}'.format(loss_train.item()),
              'acc_train: {:.4f}'.format(acc_train),
              'loss_val: {:.4f}'.format(loss_val.item()),
              'acc_val: {:.4f}'.format(acc_val),
              'time: {:.4f}s'.format(time.time() - t))


def test():
    model.eval()
    output, embeddings = model(features, adj)
    loss_test = F.nll_loss(output[idx_test], class_labels_test)
    acc_test = f1_score(torch.argmax(output[idx_test], dim=1).detach().cpu().numpy(), class_labels_test.cpu().numpy())
    print(np.sum(torch.argmax(output[idx_test], dim=1).detach().cpu().numpy()))
    print(np.sum(class_labels_test.cpu().numpy()))
    print("Test set results:",
          "loss= {:.4f}".format(loss_test.item()),
          "accuracy= {:.4f}".format(acc_test))

    return embeddings

```

```python
from sklearn.preprocessing import OneHotEncoder
from torch_geometric.data import Data

# Get a list of all relation types
relation_types = [G.edges[edge]['relation'] for edge in G.edges]

# One-hot encode the relation types
encoder = OneHotEncoder(sparse=False)
edge_features = encoder.fit_transform(np.array(relation_types).reshape(-1, 1))

# Convert edge features to tensor
edge_features = torch.tensor(edge_features, dtype=torch.float)

# Get the indices of the source and target nodes for each edge
edge_index = torch.tensor(list(G.edges), dtype=torch.long).t().contiguous()

# Get the node features and labels
node_features = torch.stack([torch.tensor(G.nodes[node]['embedding']) for node in G.nodes])
node_labels = torch.tensor([1 if G.nodes[node]['label'] == 1 else 0 for node in G.nodes], dtype=torch.long)

# Create a PyTorch Geometric data object
data = Data(x=node_features, edge_index=edge_index, edge_attr=edge_features, y=node_labels)
```

```python
data.train_mask = idx_train
data.val_mask = idx_val
data.test_mask = idx_test
```

```python
import torch
from torch.nn import Linear
from torch_geometric.nn import GCNConv
import torch.nn.functional as F
from sklearn.metrics import f1_score

class GCN(torch.nn.Module):
    def __init__(self, num_classes):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(data.num_node_features, 512)
        self.conv2 = GCNConv(512, 256)
        self.conv3 = GCNConv(256, 128)
        # self.conv4 = GCNConv(128, 32)
        self.classifier = Linear(128, num_classes)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index

        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)

        x = self.conv2(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)

        x = self.conv3(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, p=0.5, training=self.training)

        # x = self.conv4(x, edge_index)
        # x = F.relu(x)
        # x = F.dropout(x, p=0.5, training=self.training)

        x = self.classifier(x)

        return F.log_softmax(x, dim=1)
```

```python
epochs = 100
# Initialize the model and optimizer
model = GCN(num_classes=2)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
weights = [1., 3.]  # Assuming class 1 is the minority class
class_weights = torch.FloatTensor(weights).to(device)
criterion = torch.nn.NLLLoss(weight=class_weights)
# Use a GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
data = data.to(device)

# History
h = {
    'loss_train': [],
    'loss_val': [],
    'f1_train': [],
    'f1_val': []
}

# Training loop
model.train()
for epoch in trange(epochs):
    optimizer.zero_grad()
    out = model(data)
    loss = criterion(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()

    # Calculate F1-score
    model.eval()
    with torch.no_grad():
        predictions = model(data).max(dim=1)[1]
        train_f1 = f1_score(data.y[data.train_mask].cpu(), predictions[data.train_mask].cpu(), average='weighted')

    val_loss = criterion(out[data.val_mask], data.y[data.val_mask])
    val_f1 = f1_score(data.y[data.val_mask].cpu(), predictions[data.val_mask].cpu(), average='weighted')

    h['loss_train'].append(loss.item())
    h['f1_train'].append(loss.item())

    h['loss_val'].append(val_loss.item())
    h['f1_val'].append(val_f1)

    if epoch % 10 == 0:
        print(f'Epoch: {epoch+1}, Training Loss: {loss.item()}, Validation Loss: {val_loss.item()}, Training F1-score: {train_f1}, Validation F1-score: {val_f1}')

```

```python
# Calculate loss and F1-score for the test set after training
model.eval()
with torch.no_grad():
    out = model(data)
    predictions = out.max(dim=1)[1]
    test_loss = criterion(out[data.test_mask], data.y[data.test_mask])
    print(sum(predictions[data.test_mask].cpu() - data.y[data.test_mask].cpu()))
    test_f1 = f1_score(data.y[data.test_mask].cpu(), predictions[data.test_mask].cpu(), average='weighted')
    print(f1_score(data.y.cpu(), predictions.cpu(), average='weighted'))
print(f'Test Loss: {test_loss.item()}, Test F1-score: {test_f1}')
```
